```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv ('data/LOG.csv')
```

## Logs

```python
df.query("type == 0")[['time','text']]
```

## BMP280 Data

```python
bmpdf = pd.DataFrame(data=[el.split() for el in df.query("tag == 'bmp280' and type == 3").text],
                    index = df.query("tag == 'bmp280' and type == 3").time/1000,
                    columns=["temperature", "pressure"])
px.line(bmpdf)
```

## MPU9250 Data

```python
mpudf = pd.DataFrame(data=[el.split() for el in df.query("tag == 'mpu9250' and type == 4").text],
                    index = df.query("tag == 'mpu9250' and type == 4").time/1000,
                    columns=["heading", "pitch", "roll"])
px.line(mpudf)
```

## NEO6M Data

```python
neo6mdf = pd.DataFrame(data=[el.split() for el in df.query("tag == 'neo6m' and type == 5").text],
                    columns=["lat", "lon", "cog","speed","variation","valid"])
neo6mdf['lat'] = neo6mdf['lat'].astype(float)
neo6mdf['lon'] = neo6mdf['lon'].astype(float)

fig = px.line_mapbox(neo6mdf, lat="lat", lon="lon", color="speed", zoom=3, height=600)     
fig.update_layout(mapbox_style="stamen-terrain", mapbox_zoom=4, mapbox_center_lat = 41,
    margin={"r":0,"t":0,"l":0,"b":0})
fig.show()              
```
