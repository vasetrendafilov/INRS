```python
import serial
import csv
```

```python
esp32 = serial.Serial('com6', 115200)

total = 0
outputs = []
with open('mpu.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["rawX", "rawY", "rawZ", "comX","comY","comZ"])
    while total < 200:
        if total < 4:
            esp32.readline()
        else:    
             outputs.append(str(esp32.readline().decode("windows-1252")))
        total=total+1
    for output in outputs:
         writer.writerow(output.split())
esp32.close()
```
